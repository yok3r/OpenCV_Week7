```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import ImageFont, ImageDraw, Image
```

```python
def mememaker(image,text):
    memepic = cv2.imread(image)
    memepic_RGB = cv2.cvtColor(memepic,cv2.COLOR_BGR2RGB)
    
    font = ImageFont.truetype("impact.ttf", 150)
    meme_img = Image.fromarray(memepic_RGB)
    draw = ImageDraw.Draw(meme_img)

    
    #Escribim el text
    draw.text((800,1300), text, (255,255,3),font=font)
    
    
    # Imprimim la imatge
    plt.figure(figsize = (40,30))
    plt.imshow(meme_img);
    
    # Exportem el meme
    export_img = np.array(meme_img)
    export_img_2 = cv2.cvtColor(export_img,cv2.COLOR_RGB2BGR)
    cv2.imwrite("output/meme.jpg",export_img_2)
    
    return None;
```

```python
mememaker("img/meme.jpg","All I need is Fish Oil")
```
