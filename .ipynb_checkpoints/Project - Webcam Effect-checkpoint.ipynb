## Importar una imatge, identificar el numero i printarlo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#MNIST dataset

```

## Helper functions

```python
def showImg(img,title = '',plot=True,figs=(5,5)):
    img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
    if plot:
        plt.figure(figsize=figs)
        plt.title(title)
        plt.imshow(img_rgb)
    return img_rgb
    
def showGrayScaleImg(img,title = '',figs=(5,5)):
    plt.figure(figsize= figs)
    plt.title(title)
    plt.imshow(img,cmap='gray')
```

```python
image = cv2.imread("img/54.jpeg")

showImg(image, "Original image");
```

## Fem un crop de la zona:

```python
image = cv2.imread("img/54.jpeg")


image = image[570:750,250:400]


showImg(image, "Original image");
```

## La pasamos a GRIS

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Aplicamos un threshold

```python
threshold = 170
max_value = 255
```

```python
ret1, thres1 = cv2.threshold(image, threshold,max_value,cv2.THRESH_BINARY)
ret2, thres_binary_inverted = cv2.threshold(image, threshold,max_value,cv2.THRESH_BINARY_INV)
ret3, thres3 = cv2.threshold(image, threshold,max_value,cv2.THRESH_TRUNC)
ret4, thres4 = cv2.threshold(image, threshold,max_value,cv2.THRESH_TOZERO)
ret5, thres5 = cv2.threshold(image, threshold,max_value,cv2.THRESH_TOZERO_INV)
```

```python
titles = ["Original","Binary","Binary Inv", "Trunc","Tozero", "Tozero inv"]
images = [image, thres1, thres_binary_inverted, thres3, thres4, thres5]

for i in range(6):
    showGrayScaleImg(images[i],titles[i]);
```

```python
thres_binary_inverted.shape
```

## Change the size to (25,25)

```python
#Coordinades dels 4 corners
points_A = np.float32([[0,0],[150,0],[0,180],[150,180]])

#Coordinades = 25,25
points_B = np.float32([[0,0],[28,0],[0,28],[28,28]])

# Creem la matriu de la perspectiva
M = cv2.getPerspectiveTransform(points_A,points_B)


warped = cv2.warpPerspective(thres_binary_inverted, M, (28,28))

#showGrayScaleImg(M);
showGrayScaleImg(warped);
```

```python
print("Previous:" , warped.shape)

#warped_reshape = thres_binary_inverted.reshape(thres_binary_inverted, 28, 28, 1)

warped_reshape  = cv2.resize(thres_binary_inverted,(28,28),interpolation = cv2.INTER_AREA )
warped_reshape = np.reshape(warped_reshape,(1,28,28,1))

print("After reshape:" , warped_reshape.shape)
showGrayScaleImg(warped);
```

### Download the MNIST Dataset

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
image_index = 7777 
print(y_train[image_index])

plt.imshow(x_train[image_index], cmap='Greys');
```

```python
x_train.shape
```

```python
# Reshaping the array to 4-dims so that it can work with the Keras API
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)
# Making sure that the values are float so that we can get decimal points after division
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
# Normalizing the RGB codes by dividing it to the max RGB value.
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print('Number of images in x_train', x_train.shape[0])
print('Number of images in x_test', x_test.shape[0])
```

```python
# Importing the required Keras modules containing model and layers
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
# Creating a Sequential Model and adding the layers
model = Sequential()
model.add(Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten()) # Flattening the 2D arrays for fully connected layers
model.add(Dense(128, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))
```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

model.fit(x=x_train,y=y_train, epochs=2)
```

```python
model.evaluate(x_test, y_test)
```

## Prediction

```python
pred = model.predict(warped_reshape)
print("Predicted:",pred.argmax())
showGrayScaleImg(warped);
```
