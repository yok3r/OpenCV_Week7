### Helper functions

```python
def showImg(img,title = '',plot=True,figs=(10,10)):
    img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
    if plot:
        plt.figure(figsize=figs)
        plt.title(title)
        plt.imshow(img_rgb)
    return img_rgb
    
def showGrayScaleImg(img,title = '',figs=(10,10)):
    plt.figure(figsize= figs)
    plt.title(title)
    plt.imshow(img,cmap='gray')
    
all_contoruns = -1
green = (0,255,0)
thickness = 2
```

#### By area

```python
def get_contour_areas(contours):
    all_areas = []
    for contour in contours:
        contour_area = cv2.contourArea(contour)
        all_areas.append(contour_area)
    return all_areas
```

```python
madrid_contours = cv2.findContours( edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contour_areas = get_contour_areas(madrid_contours[1])
print("Number of contours:", len(madrid_contours[1]))
print(" ")
print(contour_areas)
```

### By X coordinate (left to right)

```python
def get_contour_X_center_coordinate(contours):
    if cv2.contourArea(contours) > 10:
        M = cv2.moments(contours)
        return(int(M["m10"]/M["m00"]))
```

```python
pipas = cv2.imread("img/pipas.jpeg")
showImg(pipas);

pipas_g = cv2.cvtColor(pipas, cv2.COLOR_BGR2GRAY)
gaussian = cv2.GaussianBlur(pipas_g,(3,3),-1)
#showGrayScaleImg(gaussian);
edges = cv2.Canny(gaussian,80,250)
showGrayScaleImg(edges);

pipas_contours = cv2.findContours( edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contour_areas = get_contour_areas(pipas_contours[1])
print("Number of contours:", len(pipas_contours[1]))
print(" ")
print(contour_areas)


sorted_contours = sorted(pipas_contours[1], key=cv2.contourArea ,reverse=True)
#print(sorted_contours)


cv2.drawContours(pipas,sorted_contours[:-1], -1,green,-1);
showImg(pipas);
```

```python
def sort_contours(cnts, method="left-to-right"):
    # initialize the reverse flag and sort index
    reverse = False
    i = 0

    # handle if we need to sort in reverse
    if method == "right-to-left" or method == "bottom-to-top":
        reverse = True

    # handle if we are sorting against the y-coordinate rather than
    # the x-coordinate of the bounding box
    if method == "top-to-bottom" or method == "bottom-to-top":
        i = 1

    # construct the list of bounding boxes and sort them from top to
    # bottom
    boundingBoxes = [cv2.boundingRect(c) for c in cnts]
    (cnts, boundingBoxes) = zip(*sorted(zip(cnts, boundingBoxes),
                                        key=lambda b: b[1][i], reverse=reverse))
    # return the list of sorted contours and bounding boxes
    return (cnts, boundingBoxes)
```

```python
sorted_contours , boxes = sort_contours(pipas_contours[1])
```

```python
for idx, c in enumerate(sorted_contours):
    if idx > 8:
        break
    cv2.drawContours(pipas,[c],-1,green,3)
    showImg(pipas);
    
```

### Aproximating Contourns

```python
house = cv2.imread("img/house.jpeg")

showImg(house);

house_g = cv2.cvtColor(house, cv2.COLOR_BGR2GRAY)
#house_b = cv2.GaussianBlur(house_g,(3,3),-1)
#showGrayScaleImg(house_g);

edges = cv2.Canny(house_g,80,250)

house_contours = cv2.findContours( edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print("Number of contours:", len(house_contours[1]))


for c in house_contours[1]:
    accuracy = 0.001#0.02
    perimetrer = cv2.arcLength(c, True)
    epsilon = accuracy * perimetrer
    approx = cv2.approxPolyDP(c, epsilon, True)
    
    cv2.drawContours(house, [approx], 0 , green,2)
    
    showImg(house)
    

showGrayScaleImg(edges);

```

##Fitting Rectangle

```python
hand = cv2.imread("img/hand.jpg")

hand_g = cv2.imread("img/hand.jpg",0)
ret , thresh = cv2.threshold(hand_g, 180,200, cv2.THRESH_BINARY_INV)

showGrayScaleImg(thresh);
x,y,w,h = cv2.boundingRect(thresh)

cv2.rectangle(hand,(x,y),(x+w,y+h),green,2)

showImg(hand);
```

## Fitting Circle

```python

contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

(x,y), radius = cv2.minEnclosingCircle(contours[1][0]) 

center = (int(x),int(y))
radius = int(radius)

cv2.circle(hand,center, radius, green,2)
showImg(hand);
```

## Minimum enclosing cicle

```python
coin = cv2.imread("img/coin.jpg")

coin_g = cv2.imread("img/coin.jpg",cv2.IMREAD_GRAYSCALE)
ret , thresh = cv2.threshold(coin_g, 150,200, cv2.THRESH_BINARY_INV)
contours = cv2.findContours( thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

sorted_contours = sorted(contours[1], key=cv2.contourArea ,reverse=True)
#print(sorted_contours)

#print(len(contours[1]))

(x,y), radius = cv2.minEnclosingCircle(sorted_contours[0]) 

#showGrayScaleImg(thresh);

center = (int(x),int(y))
radius = int(radius)

cv2.circle(coin,center, radius, green,5)

showImg(coin);

for i in range(len(sorted_contours)):
    (x,y), radius = cv2.minEnclosingCircle(sorted_contours[i])
    center = (int(x),int(y))
    radius = int(radius)
    cv2.circle(coin,center, radius, green,5)
showImg(coin);
```
